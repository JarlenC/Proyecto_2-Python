```python
# Generar 5 nuevas variables aplicando los siguientes criterios con las variables numéricas:
#  Log10(var1/var2)
#  Sqrt(var1)exp(var2)/200
#  Si var 1 > var 2 → 5, caso contrario 3
#  1/logn(var1/var2)*100
#  var2**2/var1

import numpy as np
```

```python
#  Log10(var1/var2)
#==========================================================================================================================================================

def obtLog10(var1, var2):
   '''La función retorna el Logaritmo en base 10 de Var1/Var2 '''
   if var2 <= 0:
     var2 = var1
   res_div = var1//var2
   return np.log10(res_div).round(2)


df_autos["NC_Log_10"] = df_autos.apply(lambda x: obtLog10(x.Peso_En_Libras, x.Caballo_De_Fuerza), axis=1)
df_autos.head(5)
```

```python
#  Sqrt(var1)exp(var2)/200
#==========================================================================================================================================================

def obtRaizQ(var1, var2):
   '''La función retorna la raíz cuadrada de Var1 elevado a Var2, dividido entre 200 '''
   resp = np.sqrt(var1**var2)//200
   return resp

df_autos["NC_RaizCua"] = df_autos.apply(lambda x: obtRaizQ(x.Aceleracion, x.Cilindros), axis=1)
df_autos.head(5)
```

```python
#  Si var 1 > var 2 → 5, caso contrario 3
#==========================================================================================================================================================

def obtMayor(var1, var2):
  '''La función retorna 5 si Var1 es mayor que Var2 , caso contrario retorna 3 '''
  if var1 > var2 :
    return 5
  return 3

df_autos["NC_Cond_5_3"] = df_autos.apply(lambda x: obtMayor(x.Millas_Por_Galon, x.Aceleracion), axis=1)
df_autos.head(11)
```

```python
#  1/logn(var1/var2)*100
#==========================================================================================================================================================

def obtLogNeper(var1, var2):
   '''La función retorna el Logaritmo Neperiano de Var1/Var2 multiplicado por 100'''
   if var2 <= 0:
     var2 = var1
   res_div = var1//var2
   return 1/(np.log(res_div)*100).round(2)

df_autos["NC_LogNep"] = df_autos.apply(lambda x: obtLogNeper(x.Peso_En_Libras, x.Cilindros), axis=1)
df_autos.head(5)
```

```python
#  var2**2/var1
#==========================================================================================================================================================

def obtElevado(var1, var2):
   '''La función retorna Var1 elevado al cuadrado dividio entre Var2'''
   if var1 <= 0:
     var1 = var2
   res_div = var2**2
   return res_div//var1

df_autos["NC_Elev2"] = df_autos.apply(lambda x: obtElevado(x.Millas_Por_Galon, x.Aceleracion), axis=1)
df_autos.head(5)
```

```python
# Generar 3 variables cualitativas ordinales con los siguientes criterios:
#  Si es mayor a la “media + 1*desviación estándar” → Alto
#-----> Nota: Mostrar con las nuevas variables generadas. Elegir con libertad las variables a trabajar.
#==========================================================================================================================================================

def esAlto(var1, varMedia):
  ''' Retorna el valor 1 en caso de cumplir la condición '''
  resp = varMedia + (1* np.std(var1))
  if var1 > resp:
    return "Si"
  else:
    return "No"

varMedia = df_autos["Caballo_De_Fuerza"].mean()
df_autos["NC_Alto"] = df_autos.apply(lambda x: esAlto(x.Caballo_De_Fuerza, varMedia), axis=1)
df_autos.tail(10)
```

```python
# Generar 3 variables cualitativas ordinales con los siguientes criterios:
#  Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
#==========================================================================================================================================================
def esMedio(var1, varMedia):
  ''' Retorna el valor 1 en caso de cumplir la condición '''
  resp1 = varMedia - (1* np.std(var1))
  resp2 = varMedia + (1* np.std(var1))
  if var1 > resp1 and var1 < resp2:
    return "Si"
  else:
    return "No"

varMedia = df_autos["Caballo_De_Fuerza"].mean()
df_autos["NC_Medio"] = df_autos.apply(lambda x: esAlto(x.Caballo_De_Fuerza, varMedia), axis=1)
df_autos.tail(10)
```

```python
# Generar 3 variables cualitativas ordinales con los siguientes criterios:
#  Si es menor a la “media - 1*desviación estándar” → Bajo
#==========================================================================================================================================================
def esBajo(var1, varMedia):
  ''' Retorna el valor 1 en caso de cumplir la condición '''
  resp = varMedia - (1* np.std(var1))
  if var1 < resp:
    return "Si"
  else:
    return "No"

varMedia = df_autos["Caballo_De_Fuerza"].mean()
df_autos["NC_Bajo"] = df_autos.apply(lambda x: esAlto(x.Caballo_De_Fuerza, varMedia), axis=1)
df_autos.tail(10)
```

```python
# Crear una función que muestre los siguientes gráficos para las variables numéricas según una variable cualitativa:
#  Cajas y bigotes
#  Barras de la media
#  Histograma con Curva

import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
import numpy as np
```

```python
#  Cajas y bigotes
#==========================================================================================================================================================
def graficos_numericos(columnas, data, x): # parámetros
  
  for i, colum in enumerate(columnas): 
    fig, ax = plt.subplots(figsize=(6, 3.84))
    sns.boxplot(   # boxplot  violinplot
        x = x,  # eje X
        y = colum,  # eje Y
        data  = data,
        )
    ax.set_title(colum, fontsize = 10, fontweight = "bold")
    ax.tick_params(labelsize = 8)
    ax.set_xlabel("País Origen USA=1, Japan=2 y Europe=3")
```

```python
x = df_autos.RC_Pais_Origen  # variable cualitativa
columnas = df_autos.select_dtypes(include=['float64', 'int']).columns
columnas
columnas = columnas.drop(["RC_Pais_Origen", "NC_Cond_5_3"])  #Generan gráficos vacíos

graficos_numericos(columnas, df_autos, x)
```

```python
#  Barras de la media
#==========================================================================================================================================================
df_media = df_autos.mean()
plt.figure()
df_media.plot.bar(stacked = False, color = "purple", label="Valor", width=0.80)
plt.legend(loc="lower left",bbox_to_anchor=(0.8,1.0))
plt.ylabel('Media')
plt.xlabel('Autos')
plt.title('Barras de la Media de autos.csv')
plt.show()
```

```python
#  Histograma con Curva
#==========================================================================================================================================================

# Valores de la media (mu) y desviación típica (sigma) de los datos
aceleracion = df_autos['Aceleracion']
mu, sigma = stats.norm.fit(aceleracion)

# Valores teóricos de la normal en el rango observado
x_hat = np.linspace(min(aceleracion), max(aceleracion), num=100)
y_hat = stats.norm.pdf(x_hat, mu, sigma)

# Gráfico
fig, ax = plt.subplots(figsize=(7,4))
ax.plot(x_hat, y_hat, linewidth=2, label='Media')
ax.hist(x=aceleracion, density=True, bins=30, color="#3182bd", alpha=0.5)
ax.plot(aceleracion, np.full_like(aceleracion, -0.01), '|k', markeredgewidth=1)
ax.set_title('Histograma con Curva de la Aceleración - Autos.csv')
ax.set_xlabel('Aceleración')
ax.set_ylabel('Densidad de probabilidad')
ax.legend();
```

```python
# Crear una función que devuelva un mapa de calor con la correlación de todas las variables numéricas (incluye las generadas), para cada país.
#==========================================================================================================================================================

def mapa_de_calor(df_autos):
  corr = df_autos.corr(method='pearson')
  fig, ax = plt.subplots(figsize=(10,10))
  im = ax.imshow(corr, interpolation='nearest')
  fig.colorbar(im, orientation='vertical', fraction = 0.05)

  ax.set_xticklabels(df_autos.columns, rotation=65, fontsize=15)
  ax.set_yticklabels(df_autos.columns, rotation=0, fontsize=15)
  
  for i in range(len(df_autos.columns)-18):
      for j in range(len(df_autos.columns)-18):
          text = ax.text(j, i, round(corr.to_numpy()[i, j], 2), ha="center", va="center", color="black")

  plt.show()  

mapa_de_calor(df_autos)
```
